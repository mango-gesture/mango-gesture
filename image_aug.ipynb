```python
from PIL import Image
```

```python
# load 0.jpeg and 1.jpeg
img1 = Image.open(f'test_data/0/output0.jpg')
img2 = Image.open(f'test_data/0/output1.jpg')
#display both in matplotlib
import matplotlib.pyplot as plt
plt.figure()
plt.subplot(1,2,1)
plt.imshow(img1)
plt.subplot(1,2,2)
plt.imshow(img2)
```

```python
from scipy.ndimage import gaussian_filter

# apply a gaussian filter to both images
img1_gauss = gaussian_filter(img1, sigma=2)
img2_gauss = gaussian_filter(img2, sigma=2)

# display both images using matplotlib
plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(img1_gauss)
plt.subplot(1, 2, 2)
plt.imshow(img2_gauss)
```

```python
# Convert numpy arrays to PIL images
img1_gauss_bw = Image.fromarray(img1_gauss).convert('L')
img2_gauss_bw = Image.fromarray(img2_gauss).convert('L')

# Display both images using matplotlib
plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(img1_gauss_bw, cmap='gray')
plt.subplot(1, 2, 2)
plt.subplot(1,2,2)
plt.imshow(img2_gauss_bw, cmap='gray')
```

```python
# add gaussian noise to both original images
import numpy as np
img1 = np.array(img1)
img2 = np.array(img2)
img1_gauss = img1 + np.random.normal(0, 40, img1.shape)
#normalize
img1_gauss = np.clip(img1_gauss, 0, 255).astype(np.uint8)
img2_gauss = img2 + np.random.normal(0, 40, img2.shape)
#normalize
img2_gauss = np.clip(img2_gauss, 0, 255).astype(np.uint8)
#display
plt.figure()
plt.subplot(1,2,1)
plt.imshow(img1_gauss)
plt.subplot(1,2,2)
plt.imshow(img2_gauss)

```
